```python
import pandas as pd
```

```python
news_data = pd.read_csv("financialNews.csv")
sentiment_data = pd.read_csv("sentiment_analysis_results.csv")
```

```python
news_data.head()
```

```python
print(sentiment_data)
```

```python
# Merge sentiment data with stock data
# Convert Date and datetime columns to consistent datetime format
news_data['Date'] = pd.to_datetime(news_data['datetime'])
sentiment_data['datetime'] = pd.to_datetime(sentiment_data['datetime'])
```

```python
# Merge the datasets based on the 'title' column
merged_data = news_data.merge(sentiment_data[['title', 'sentiment']], 
                                  on='title', 
                                  how='left') 
```

```python
merged_data.head()
```

```python
merged_data = merged_data.drop(columns= ['Date', 'change_pct'])
```

```python
# Save the result to a new CSV if needed
merged_data.to_csv('financialNews_with_sentiment.csv', index=False)
```
